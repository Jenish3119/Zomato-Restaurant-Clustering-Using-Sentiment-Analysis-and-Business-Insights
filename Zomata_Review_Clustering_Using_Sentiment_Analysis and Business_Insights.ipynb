# **Project Summary -**

The objective of this project is to analyze customer reviews and restaurant information from the Zomato dataset to generate meaningful business insights and segment restaurants into distinct groups using unsupervised machine learning techniques. The dataset consists of restaurant metadata, customer reviews, ratings, cuisine information, restaurant costs, review timestamps, and customer engagement features. By combining Natural Language Processing (NLP), Exploratory Data Analysis (EDA), feature engineering, and clustering algorithms, the project aims to understand customer preferences and support data-driven business decisions.

The project began with comprehensive data preprocessing and cleaning. Missing values, duplicate records, and inconsistent entries were handled appropriately to improve data quality. The textual review data underwent several NLP preprocessing steps, including lowercasing, contraction expansion, punctuation removal, URL removal, stopword removal, tokenization, and lemmatization. Sentiment analysis was then performed using the VADER sentiment analyzer to generate sentiment scores and classify customer reviews as positive, neutral, or negative.

Extensive Exploratory Data Analysis (EDA) was carried out to understand the distribution of ratings, restaurant costs, cuisines, review counts, and customer engagement. Multiple visualizations, including histograms, bar charts, scatter plots, box plots, heatmaps, word clouds, and multivariate plots, were used to identify trends and relationships among variables. Statistical hypothesis testing further validated relationships between restaurant cost, ratings, and customer behavior.

To prepare the data for clustering, review-level data was transformed into restaurant-level features through aggregation. New features such as average rating, average sentiment score, review count, average review length, average word count, cuisine diversity, picture uploads, and positive and negative review percentages were engineered. These features provided a comprehensive representation of each restaurant's performance and customer engagement.

The numerical features were standardized using StandardScaler, followed by Principal Component Analysis (PCA) for dimensionality reduction and visualization. Three unsupervised clustering algorithms—K-Means, Agglomerative Clustering, and DBSCAN—were implemented to segment restaurants based on their characteristics. The optimal number of clusters for K-Means was determined using the Elbow Method and Silhouette Score. Model performance was evaluated using the Silhouette Score, which measures cluster cohesion and separation.

Among the three algorithms, DBSCAN achieved the best performance with a Silhouette Score of 0.6922, outperforming Agglomerative Clustering (0.3340) and K-Means (0.3310). The resulting clusters revealed meaningful restaurant segments based on pricing, customer satisfaction, review sentiment, customer engagement, and cuisine diversity. These clusters can support personalized restaurant recommendations, targeted marketing campaigns, pricing optimization, customer retention strategies, and identification of premium or underperforming restaurants.

Overall, this project demonstrates the practical application of NLP, sentiment analysis, feature engineering, clustering algorithms, and business analytics to solve a real-world restaurant segmentation problem. The insights generated from this analysis can help restaurant aggregators such as Zomato better understand customer preferences, improve recommendation systems, and make informed business decisions through data-driven segmentation.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# Remove Duplicate Records
df_reviews = df_reviews.drop_duplicates()

# Handle Missing Values
df_meta["Collections"] = df_meta["Collections"].fillna("Not Available")
df_meta["Timings"] = df_meta["Timings"].fillna("Not Available")

df_reviews = df_reviews.dropna()

# Clean Rating Column
df_reviews["Rating"] = df_reviews["Rating"].replace("Like", np.nan)
df_reviews["Rating"] = pd.to_numeric(df_reviews["Rating"], errors="coerce")
df_reviews = df_reviews.dropna(subset=["Rating"])

# Merge Both Datasets
df = pd.merge(
    df_reviews,
    df_meta,
    left_on="Restaurant",
    right_on="Name",
    how="left"
)

print("Final Dataset Shape:", df.shape)
df.head()
```

### What all manipulations have you done and insights you found?

Removed 36 duplicate records from the reviews dataset.

Filled missing values in the Collections and Timings columns of the metadata dataset.

Removed rows with missing values from the reviews dataset to ensure data quality.

Converted the Rating column from object to numeric format and removed invalid values.

Merged the restaurant metadata and reviews datasets using the restaurant name to create a unified dataset for analysis.

The final dataset combines restaurant details with customer reviews, making it suitable for exploratory data analysis and machine learning.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Does the restaurant cost affect customer ratings?

Null Hypothesis (H₀) - There is no significant difference in customer ratings across different cost categories.

Alternative Hypothesis (H₁) - There is a significant difference in customer ratings across different cost categories.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

low = df[df["Cost Category"]=="Low"]["Rating"]
medium = df[df["Cost Category"]=="Medium"]["Rating"]
high = df[df["Cost Category"]=="High"]["Rating"]
premium = df[df["Cost Category"]=="Premium"]["Rating"]

f_stat, p_value = f_oneway(low, medium, high, premium)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA (Analysis of Variance)

##### Why did you choose the specific statistical test?

One-Way ANOVA is appropriate because we are comparing the mean customer ratings across more than two independent cost categories (Low, Medium, High, and Premium). It determines whether at least one group's mean rating differs significantly from the others.

#### 2. Cross- Validation & Hyperparameter Tuning

```python
# Tune eps and min_samples using silhouette score

dbscan_results = []

eps_values = [0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]
min_samples_values = [3, 4, 5, 6, 7]

for eps in eps_values:
    for min_samples in min_samples_values:

        model = DBSCAN(
            eps=eps,
            min_samples=min_samples
        )

        labels = model.fit_predict(X_scaled)

        # Exclude noise points for silhouette calculation
        valid_mask = labels != -1
        valid_labels = labels[valid_mask]
        valid_data = X_scaled[valid_mask]

        n_clusters = len(set(valid_labels))

        # Silhouette requires at least two clusters
        if n_clusters >= 2 and len(valid_data) > n_clusters:
            score = silhouette_score(
                valid_data,
                valid_labels
            )

            dbscan_results.append({
                "eps": eps,
                "min_samples": min_samples,
                "clusters": n_clusters,
                "noise_points": np.sum(labels == -1),
                "silhouette_score": score
            })

dbscan_tuning_df = pd.DataFrame(dbscan_results)

display(
    dbscan_tuning_df.sort_values(
        "silhouette_score",
        ascending=False
    ).head(10)
)
```

```python
# Select the best DBSCAN configuration

best_dbscan = (
    dbscan_tuning_df
    .sort_values(
        "silhouette_score",
        ascending=False
    )
    .iloc[0]
)

best_eps = best_dbscan["eps"]
best_min_samples = int(best_dbscan["min_samples"])

print("Best eps:", best_eps)
print("Best min_samples:", best_min_samples)
print(
    "Best Silhouette Score:",
    round(best_dbscan["silhouette_score"], 4)
)
```

```python
# Fit DBSCAN using optimized parameters

dbscan_final = DBSCAN(
    eps=best_eps,
    min_samples=best_min_samples
)

restaurant_df["DBSCAN_Cluster"] = (
    dbscan_final.fit_predict(X_scaled)
)

print(
    restaurant_df["DBSCAN_Cluster"]
    .value_counts()
    .sort_index()
)
```

```python
# Evaluate tuned DBSCAN

dbscan_labels = restaurant_df["DBSCAN_Cluster"].values

valid_mask = dbscan_labels != -1
valid_labels = dbscan_labels[valid_mask]
valid_data = X_scaled[valid_mask]

if len(set(valid_labels)) >= 2:
    dbscan_score = silhouette_score(
        valid_data,
        valid_labels
    )

    plt.figure(figsize=(5, 4))
    plt.bar(
        ["DBSCAN"],
        [dbscan_score]
    )

    plt.ylim(0, 1)
    plt.ylabel("Silhouette Score")
    plt.title("DBSCAN Evaluation")
    plt.show()

    print(
        "DBSCAN Silhouette Score:",
        round(dbscan_score, 4)
    )

else:
    dbscan_score = np.nan
    print(
        "DBSCAN produced fewer than two valid clusters, "
        "so the silhouette score cannot be calculated."
    )
```
